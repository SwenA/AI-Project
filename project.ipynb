```python
import numpy as np
import matplotlib
import pandas as pd
import matplotlib.pyplot as plt
import zipfile
import warnings
from numpy.random import default_rng

random=default_rng(42) 

warnings.filterwarnings('ignore')
pd.set_option('display.max_columns', None)
general_data = pd.read_csv('./data/general_data.csv')
general_data.info()
```

```python
manager_survey_data = pd.read_csv('./data/manager_survey_data.csv')
manager_survey_data.info()
```

```python
employee_survey_data = pd.read_csv('./data/employee_survey_data.csv')
employee_survey_data.info()
```

```python
with zipfile.ZipFile('./data/in_out_time.zip', 'r') as z:
    with z.open('in_time.csv') as i:
        in_time_data = pd.read_csv(i)
    with z.open('out_time.csv') as o:
        out_time_data = pd.read_csv(o)

in_time_data.head()
```

```python
print(type(in_time_data.iloc[1,1]))
in_time=in_time_data.replace(np.nan,0)
print(type(in_time.iloc[1,4]))
in_time.head()
```

```python
in_time.iloc[:, 1:] = in_time.iloc[:, 1:].apply(pd.to_datetime, errors='coerce', format='mixed')
in_time.head()
```

```python
type(in_time.iloc[1, 4])
```

```python
out_time=out_time_data.replace(np.nan,0)
```

```python
out_time.iloc[:, 1:] = out_time.iloc[:, 1:].apply(pd.to_datetime, errors='coerce', format='mixed')
print(out_time.iloc[1,4])
out_time.head()
```

```python
type(out_time.iloc[1, 4])
```

```python
worked_time = out_time - in_time
worked_time.reset_index(inplace=True)
print(worked_time.info())
worked_time.head()
```

```python
type(worked_time.iloc[1,3])
```

```python
worked_time.drop(columns=['index','Unnamed: 0'],axis=1,inplace=True)
worked_time.head()
```

```python
columns_to_delete = ['2015-01-01', '2015-01-14', '2015-01-26', '2015-03-05',
                        '2015-05-01', '2015-07-17', '2015-09-17', '2015-10-02',
                        '2015-11-09', '2015-11-10', '2015-11-11', '2015-12-25']

for col in columns_to_delete:
    if col in worked_time.columns:
        worked_time.drop([col], axis=1, inplace=True)

worked_time.info()
```

```python
print(type(worked_time.iloc[1,0]))
```

```python
worked_time_seconds = worked_time.applymap(lambda x: x.total_seconds())
worked_time['Actual Time']=worked_time_seconds.mean(axis=1)
worked_time['Actual Time']=pd.to_timedelta(worked_time['Actual Time'], unit='s')
worked_time['Actual Time'].head()
```

```python
worked_time['hrs']=worked_time['Actual Time']/np.timedelta64(1, 'h')
worked_time.head()
```

```python
worked_time.reset_index(inplace=True)
worked_time.head()
```

```python
worked_time.drop(worked_time.columns.difference(['index','hrs']), axis=1, inplace=True)
```

```python
worked_time.rename(columns={'index': 'EmployeeID'},inplace=True)
worked_time.head()
```

```python
df_1 = pd.merge(employee_survey_data, general_data, how='inner', on='EmployeeID')
all_data = pd.merge(manager_survey_data, df_1, how='inner', on='EmployeeID')
all_data = pd.merge(worked_time, all_data, how='inner', on='EmployeeID')
all_data.head()
```

```python
from sklearn.preprocessing import OneHotEncoder

label_cat = all_data[["Attrition"]]

cat_encoder = OneHotEncoder()
label_cat_1hot=cat_encoder.fit_transform(label_cat)
label_cat_1hot # array creux par dfaut

```

```python
all_data['Attrition'] = label_cat_1hot.toarray()
```

```python
all_data.head()
all_data['Attrition']
```

```python
from sklearn.model_selection import train_test_split

train_set, test_set = train_test_split(all_data)
```

```python
X_train = train_set.drop("Attrition", axis=1)
X_test = test_set.drop("Attrition", axis=1)
y_train = train_set["Attrition"].copy()
y_test = test_set["Attrition"].copy()

from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.compose import ColumnTransformer

num_pipeline = Pipeline([
        ('imputer', SimpleImputer(strategy="median")),
        ('std_scaler', StandardScaler()),
    ])

X_num = X_train.select_dtypes(include=[np.number])
X_cat = X_train.select_dtypes(include=['object', 'string']) 
num_attribs = list(X_num)
cat_attribs = list(X_cat)

full_pipeline = ColumnTransformer([
        ("num", num_pipeline, num_attribs),
        ("cat", OneHotEncoder(), cat_attribs),
    ])

X_prepared = full_pipeline.fit_transform(X_train)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import cross_val_predict


lgt_clf = LogisticRegression(random_state=42)
y_scores = cross_val_predict(lgt_clf, X_prepared, y_train, cv=3, method='decision_function')
```

```python
from sklearn.metrics import roc_curve

fpr, tpr, thresholds = roc_curve(y_train, y_scores)
```

```python
def plot_roc_curve(fpr, tpr, label=None):
    plt.plot(fpr, tpr, linewidth=2, label=label)
    plt.plot([0, 1], [0, 1], 'k--')
    plt.axis([0, 1, 0, 1])
    plt.xlabel('False Positive Rate', fontsize=16)
    plt.ylabel('True Positive Rate', fontsize=16)
```

```python
plt.figure(figsize=(8, 6))
plot_roc_curve(fpr, tpr, "Logistic Regression")
plt.show()
```

```python
from sklearn.metrics import roc_auc_score

roc_auc_score(y_train, y_scores)
```

```python
X_test_prepared = full_pipeline.fit_transform(X_test)
lgt_clf = LogisticRegression(random_state=42)
y_scores = cross_val_predict(lgt_clf, X_test_prepared, y_test, cv=3, method='decision_function')
fpr, tpr, thresholds = roc_curve(y_test, y_scores)
plt.figure(figsize=(8, 6))
plot_roc_curve(fpr, tpr, "Logistic Regression")
plt.show()
```

```python
full_pipeline_with_predictor = Pipeline([
        ('preprocessing', full_pipeline),
        ('predictor', lgt_clf)
    ])

full_pipeline_with_predictor.fit(X_train, y_train)
predictions = full_pipeline_with_predictor.predict(X_test)

fpr, tpr, thresholds = roc_curve(y_test, predictions)

plt.figure(figsize=(8, 6))
plot_roc_curve(fpr, tpr, "Logistic Regression")
plt.show()
```
