```python
import numpy as np
import matplotlib
import pandas as pd
import matplotlib.pyplot as plt
import zipfile
import warnings
from numpy.random import default_rng

random=default_rng(42) 

warnings.filterwarnings('ignore')
pd.set_option('display.max_columns', None)
general_data = pd.read_csv('./data/general_data.csv')
general_data.info()
```

```python
manager_survey_data = pd.read_csv('./data/manager_survey_data.csv')
manager_survey_data.info()
```

```python
employee_survey_data = pd.read_csv('./data/employee_survey_data.csv')
employee_survey_data.info()
```

```python
with zipfile.ZipFile('./data/in_out_time.zip', 'r') as z:
    with z.open('in_time.csv') as i:
        in_time = pd.read_csv(i)
    with z.open('out_time.csv') as o:
        out_time = pd.read_csv(o)
```

```python
in_time.head()
```

```python
in_time.iloc[:, 1:] = in_time.iloc[:, 1:].apply(pd.to_datetime)
in_time.head()
```

```python
out_time.iloc[:, 1:] = out_time.iloc[:, 1:].apply(pd.to_datetime)
out_time.head()
```

```python
worked_time = out_time.iloc[:,1:] - in_time.iloc[:,1:]
```

```python
worked_time.head()
```

```python
worked_time = worked_time/np.timedelta64(1, 'h')
worked_time['hrs'] = worked_time.mean(axis=1)
```

```python
worked_time.head()
```

```python
worked_time.reset_index(inplace=True)
worked_time['index'] += 1
```

```python
worked_time.head()
```

```python
worked_time.drop(worked_time.columns.difference(['index','hrs']), axis=1, inplace=True)
```

```python
worked_time.rename(columns={'index': 'EmployeeID'},inplace=True)
```

```python
worked_time.head(6)
```

```python
df_1 = pd.merge(employee_survey_data, general_data, how='inner', on='EmployeeID')
all_data = pd.merge(manager_survey_data, df_1, how='inner', on='EmployeeID')
all_data = pd.merge(worked_time, all_data, how='inner', on='EmployeeID')
all_data = all_data.drop(['EmployeeID'], axis=1)
```

```python
all_data.head()
```

```python
from sklearn.preprocessing import OneHotEncoder

label_cat = all_data[["Attrition"]]

cat_encoder = OneHotEncoder(categories=[['No', 'Yes']], drop='if_binary')
label_cat_1hot=cat_encoder.fit_transform(label_cat)
label_cat_1hot

```

```python
all_data['Attrition'] = label_cat_1hot.toarray()
```

```python
all_data['Attrition']
```

```python
all_data['Education'] = all_data['Education'].replace({ 1 : 'Below College', 2: 'College',3: 'Bachelor',4: 'Master',5 : 'PhD'})
all_data['EnvironmentSatisfaction'] = all_data['EnvironmentSatisfaction'].replace({ 1 : 'Low', 2: 'Medium',3: 'High',4: 'Very High'})
all_data['JobInvolvement'] = all_data['JobInvolvement'].replace({ 1 : 'Low', 2: 'Medium',3: 'High',4: 'Very High'})
all_data['JobSatisfaction'] = all_data['JobSatisfaction'].replace({ 1 : 'Low', 2: 'Medium',3: 'High',4: 'Very High'})
all_data['PerformanceRating'] = all_data['PerformanceRating'].replace({ 1 : 'Low', 2: 'Good',3: 'Excellent',4: 'Outstanding'})
all_data['WorkLifeBalance'] = all_data['WorkLifeBalance'].replace({ 1 : 'Bad', 2: 'Good',3: 'Better',4: 'Best'})
```

```python
all_data.head()
```

```python
all_data = all_data.drop(['EmployeeCount', 'Over18', 'StandardHours'], axis=1)
```

```python
from sklearn.model_selection import train_test_split

def splitting_data(data):    
    train_set, test_set = train_test_split(all_data, random_state=42)
    X_train = train_set.drop("Attrition", axis=1)
    X_test = test_set.drop("Attrition", axis=1)
    y_train = train_set["Attrition"].copy()
    y_test = test_set["Attrition"].copy()

    return X_train, X_test, y_train, y_test
```

```python
X_train, X_test, y_train, y_test = splitting_data(all_data)
```

```python
def splitting_num_cat(X):
    X_num = X.select_dtypes(include=[np.number])
    X_cat = X.select_dtypes(include=['object', 'string']) 
    num_attribs = list(X_num)
    cat_attribs = list(X_cat)

    return num_attribs, cat_attribs
```

```python
num_attribs, cat_attribs = splitting_num_cat(X_train)
```

```python
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.compose import ColumnTransformer
        
num_pipeline = Pipeline([
        ('imputer', SimpleImputer(strategy="median")),
        ('std_scaler', StandardScaler()),
    ])

full_pipeline = ColumnTransformer([
        ("num", num_pipeline, num_attribs),
        ("cat", OneHotEncoder(sparse_output=True), cat_attribs),
    ],
    remainder='passthrough')

X_prepared = full_pipeline.fit_transform(X_train)
```

```python
def plot_roc_curve(fpr, tpr, label=None):
    plt.plot(fpr, tpr, linewidth=2, label=label)
    plt.plot([0, 1], [0, 1], 'k--')
    plt.axis([0, 1, 0, 1])
    plt.xlabel('False Positive Rate', fontsize=16)
    plt.ylabel('True Positive Rate', fontsize=16)
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_curve, roc_auc_score, confusion_matrix
from sklearn.model_selection import cross_val_predict

def cross_validation_predict (model, X, y, cv=3, method=None, label=None, retrn=True):

    if not isinstance(method, str):
        raise ValueError("Le paramètre 'method' doit être une chaîne de caractères (string).")
    
    if not isinstance(label, str):
        raise ValueError("Le paramètre 'label' doit être une chaîne de caractères (string).")
    
    if  method == 'decision_function':
        cvp_scores = cross_val_predict(model, X, y, cv=cv, method='predict', scoring='accuracy')

        print('accuracy', accuracy_score(y, cvp_scores))
        print('precision: ', precision_score(y, cvp_scores))
        print('recall: ', recall_score(y, cvp_scores))
        print('f1 score: ', f1_score(y, cvp_scores))
        print('Confusion matrix: \n', confusion_matrix(y, cvp_scores))


        cvp_scores = cross_val_predict(model, X, y, cv=cv, method=method)
        fpr, tpr, thresholds = roc_curve(y,  cvp_scores)

        print('AUC: ',roc_auc_score(y, cvp_scores))

        plt.figure(figsize=(8, 6))
        plt.title(label + ' ROC curve')
        plot_roc_curve(fpr, tpr, label)
        plt.show()

        if retrn:
            return fpr, tpr
    

    if method == 'predict_proba':
        cvp_scores = cross_val_predict(model, X, y, cv=cv, method='predict', scoring='accuracy')

        print('accuracy', accuracy_score(y, cvp_scores))
        print('precision: ', precision_score(y, cvp_scores))
        print('recall: ', recall_score(y, cvp_scores))
        print('f1 score: ', f1_score(y, cvp_scores))
        print('Confusion matrix: \n', confusion_matrix(y, cvp_scores))


        cvp_scores = cross_val_predict(model, X, y, cv=cv, method=method)
        fpr, tpr, thresholds = roc_curve(y,  cvp_scores[:,1])

        print('AUC: ',roc_auc_score(y, cvp_scores[:,1]))
        
        plt.figure(figsize=(8, 6))
        plt.title(label + ' ROC curve')
        plot_roc_curve(fpr, tpr, label)
        plt.show()

        if retrn:
            return fpr, tpr
```

```python
from sklearn.linear_model import SGDClassifier
from sklearn.model_selection import RandomizedSearchCV
from scipy.stats import randint


sgd_clf = SGDClassifier(random_state=42)

param_distribs_sgd = {
    'max_iter': randint(500, 5000),
    'loss': ['hinge', 'log_loss', 'modified_huber', 'squared_hinge', 'perceptron', 'squared_error', 'huber', 'epsilon_insensitive', 'squared_epsilon_insensitive'],
    'penalty': ['l2', 'l1', 'elasticnet', None],
    'tol': [0.0001],
}

sgd_rnd_search_cv = RandomizedSearchCV(sgd_clf, param_distribs_sgd, n_iter=100, cv=3, scoring='accuracy', random_state=42, verbose=2)

sgd_rnd_search_cv.fit(X_prepared, y_train)

print("Meilleurs paramètres:", sgd_rnd_search_cv.best_params_)
print("Meilleur score:", sgd_rnd_search_cv.best_score_)

best_sgd_clf = sgd_rnd_search_cv.best_estimator_
```

```python
lgr_fpr, lgr_tpr = cross_validation_predict(best_sgd_clf, X_prepared, y_train, method='predict_proba', label='Logistic regression')
```

```python
from sklearn.linear_model import LogisticRegression



lgr_clf = LogisticRegression(random_state=42)

param_distribs_lgr = {
    'max_iter': (100, 200, 500, 1000),
    'penalty': ['l1', 'l2', 'elasticnet', None],
    'solver': ['lbfgs', 'liblinear', 'newton-cg', 'newton-cholesky'],
    'multi_class': ['ovr', 'multinomial']
}

lgr_rnd_search_cv = RandomizedSearchCV(lgr_clf, param_distribs_lgr, n_iter=80, cv=5, scoring='accuracy', random_state=42, verbose=2)

lgr_rnd_search_cv.fit(X_prepared, y_train)

print("Meilleurs paramètres:", lgr_rnd_search_cv.best_params_)
print("Meilleur score:", lgr_rnd_search_cv.best_score_)

best_lgr_clf = lgr_rnd_search_cv.best_estimator_
```

```python
lgr_fpr, lgr_tpr = cross_validation_predict(best_lgr_clf, X_prepared, y_train, method='predict_proba', label='Logistic regression')
```

```python
from sklearn.neural_network import MLPClassifier

mlp_clf = MLPClassifier(max_iter=500, random_state=42)

param_distribs_mlp = {
    'hidden_layer_sizes': [(50,), (100,), (50, 50), (100, 50)],
    'activation': ['identity', 'logistic', 'relu', 'tanh'],
    'solver': ['lbfgs','adam', 'sgd'],
    'alpha': [0.0001, 0.001, 0.01],
    'learning_rate_init': [0.001, 0.01, 0.1],
    'learning_rate': ['constant', 'invscaling', 'adaptive']
}

mlp_rnd_search_cv = RandomizedSearchCV(mlp_clf, param_distribs_mlp, n_iter=30, cv=10, scoring='accuracy', random_state=42, verbose=2)

mlp_rnd_search_cv.fit(X_prepared, y_train)

print("Meilleurs paramètres:", mlp_rnd_search_cv.best_params_)
print("Meilleur score:", mlp_rnd_search_cv.best_score_)

best_mlp_clf = mlp_rnd_search_cv.best_estimator_
```
